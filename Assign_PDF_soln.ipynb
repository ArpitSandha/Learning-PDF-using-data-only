```python
import pandas as pd
import numpy as np

df= pd.read_csv("/content/data.csv", encoding='latin1', low_memory=False)
print(df.columns)

x=df["no2"].dropna()
x=pd.to_numeric(x,errors='coerce')
x=x[x>=0]
x=x.values

r=102303798
ar= 0.05*(r%7)
br=0.3+(r%5+1)
print("ar=", ar)
print('br=',br)

z=x+ ar*np.sin(br*x)
print('First 30 transformed values=',z[:30])

mu=np.mean(z)
print('mu=',mu)

var=np.var(z)
print('variance=',var)

lam= 1/(2*var)
print("Lambda=",lam)

c= 1/np.sqrt(2*np.pi*var)
print('c=',c)
```
